Step 1: Import Libraries and Load the Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (replace 'student_performance.csv' with your actual file path)
df = pd.read_csv('https://raw.githubusercontent.com/Lohi28/google-colab/refs/heads/main/StudentsPerformance.csv')

# Quick look at the data
print(df.head())
```

```python
print(df.head())  # Display the first few rows of the dataset
```

```python
print(df.columns)
```

Step 2: Histogram of Scores

```python
# Create a histogram to show the distribution of final scores
plt.figure(figsize=(8, 6))
sns.histplot(df['math score'], kde=True, bins=10, color='blue') # Changed column name to 'math score'
plt.title('Distribution of Math Scores') # Changed title to reflect column name change
plt.xlabel('Math Score') # Changed x-axis label to reflect column name change
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Step 3: Boxplot by Gender

```python
# Create a boxplot to compare math scores by gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='math score', data=df, palette='Set2')  # Adjust 'math score' if needed
plt.title('Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.grid(True)
plt.show()

```

Step 4: Correlation Heatmap

```python
# Generate a correlation matrix
plt.figure(figsize=(10, 8))
corr_matrix = df.corr(numeric_only=True) # Add numeric_only=True

# Create a heatmap for correlations between numeric variables
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

```

Step 5: Scatterplot


```python
# prompt: generate scatter plot for 'parental level of education', 'lunch',
#        'test preparation course', 'math score'

# Step 5: Scatterplot of Study Hours vs. Final Scores
plt.figure(figsize=(8, 6))
sns.scatterplot(x='parental level of education', y='math score', hue='lunch', data=df)
plt.title('Math Score vs Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Math Score')
plt.grid(True)
plt.show()


plt.figure(figsize=(8, 6))
sns.scatterplot(x='test preparation course', y='math score', hue='lunch', data=df)
plt.title('Math Score vs Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Math Score')
plt.grid(True)
plt.show()
```

Step 6: Bar Chart

```python
# prompt: generate barplot for 'gender', 'race/ethnicity', 'reading score',
#        'writing score'

# Step 6: Barplot for Gender, Race/Ethnicity, Reading Score, Writing Score

# Barplot for Gender and Reading Score
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='reading score', data=df, palette='Set3')
plt.title('Average Reading Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Reading Score')
plt.show()

# Barplot for Race/Ethnicity and Writing Score
plt.figure(figsize=(10, 6))
sns.barplot(x='race/ethnicity', y='writing score', data=df, palette='Set3')
plt.title('Average Writing Score by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Writing Score')
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.show()
```
