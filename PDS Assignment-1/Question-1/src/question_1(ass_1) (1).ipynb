```python
# prompt: read csv

import pandas as pd

# Assuming your CSV file is named 'data.csv' and is in your current working directory
df = pd.read_csv('https://raw.githubusercontent.com/Lohi28/google-colab/refs/heads/main/Data.csv')

# Print the first 5 rows of the DataFrame
print(df.head())
```

```python
# prompt: print columns and strip space

print(df.columns)
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
# prompt: print frailty convert to 1 and 0
print(df['Frailty'].unique())


df['Frailty'] = df['Frailty'].map({' Y': 1, ' N': 0})

print(df.head())
```

```python
# prompt: generate cleaned df and save

# Remove rows with missing values (if any)
df_cleaned = df.dropna()

# Save the cleaned DataFrame to a CSV file
df_cleaned.to_csv('cleaned_data.csv', index=False)

# Print a message to confirm the save
print("Cleaned DataFrame saved to 'cleaned_data.csv'")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatterplot for Age vs Grip strength
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Grip strength', data=df, hue='Frailty', style='Frailty', palette='Set1')
plt.title('Age vs Grip Strength (with Frailty)')
plt.xlabel('Age (Years)')
plt.ylabel('Grip Strength (kg)')
plt.grid(True)
plt.show()
```

```python
# Boxplot for Grip Strength by Frailty
plt.figure(figsize=(8, 6))
sns.boxplot(x='Frailty', y='Grip strength', data=df, palette='Set2')
plt.title('Grip Strength by Frailty')
plt.xlabel('Frailty')
plt.ylabel('Grip Strength (kg)')
plt.grid(True)
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Height', 'Weight', 'Age', 'Grip strength']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# prompt: generate barplot for above data

plt.figure(figsize=(8, 6))
sns.countplot(x='Frailty', data=df)
plt.title('Distribution of Frailty')
plt.xlabel('Frailty')
plt.ylabel('Count')
plt.show()
```

```python
# prompt: generate histogram for age and grip strength at a time

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df['Grip strength'], bins=10, edgecolor='black')
plt.title('Histogram of Grip Strength')
plt.xlabel('Grip Strength')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
